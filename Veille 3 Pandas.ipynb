# Regroupement et tri

```python
import pandas as pd
reviews = pd.read_csv("C:/Users/user/Desktop/ODC/covid.csv", index_col=0)
pd.set_option("display.max_rows", 15)
reviews
```

```python
reviews.groupby('Population').Population.count()
```

```python
reviews.groupby('Population').TotalCases.min()
```

```python
reviews.groupby('Continent').apply(lambda df: df.TotalCases.iloc[0])
```

```python
reviews.groupby(['Country/Region', 'Continent']).apply(lambda df: df.loc[df.Population.idxmax()])
```

```python
reviews.groupby(['continent']).Population.agg([len, min, max])
```

```python
countries_reviewed = reviews.groupby(['Country/Region', 'Continent']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```
